```python
import pandas as pd
import numpy as np
from apyori import apriori  
```

```python
data = pd.read_csv('store_data.csv', header = None)
data.head()
```

```python
#Apyori needs the dataframe to be preprocessed in the format below 

records = []  
for i in range(0, 7501):  
    records.append([str(data.values[i,j]) for j in range(0, 20)])
records[0:4]
```

```python
#create the model to recommend products that are usually bought together 
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=4)  
association_results = list(association_rules)  
```

```python
association_results[0:5]
```

```python
#print out the a couple examples of the association rules

for  i, item in enumerate(association_results):
    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
        
    if i ==3:
        break
```
